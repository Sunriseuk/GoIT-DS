```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Частина перша: Знайомство з Pandas

```python
df = pd.read_html("https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C",
                   attrs={"class": "wikitable collapsible collapsed", "style": "text-align: right"}, match="Регіон",
                   decimal=',', thousands='.')
df = df[0]
df
```

Вивести перші рядки таблиці за допомогою методу head

```python
df.head(5)
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace({"—": np.nan})
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові

```python
df['2014'] = pd.to_numeric(df['2014'])
df['2019'] = pd.to_numeric(df['2019'])
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum()/df.shape[0]
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(axis=0, index=27, inplace=True)
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = df.fillna({"1950": df["1950"].mean(), "1960": df["1960"].mean(), "1970": df["1970"].mean(),
                "1990": df["1990"].mean(), "2000": df["2000"].mean(), "2012": df["2012"].mean(),
                "2014": df["2014"].mean(), "2019": df["2019"].mean()})
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df[df["2019"] > df["2019"].mean()][["Регіон", "2019"]].sort_values("2019")
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df.sort_values(["2014"], ascending=False).head(1)[["Регіон", "2014"]]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
fig, ax = plt.subplots(figsize=(16, 10))
df.plot(x="Регіон", y="2019", kind="bar",
        fontsize=14, color='#9966FF', ax=ax, grid=True)
plt.title("Діаграма народжуваності по регіонах у 2019 році", fontdict = {'fontsize' : 18})
plt.ylim(0, 12)
ax.set_xlabel("Регіон", size=16, color="blue")
ax.set_ylabel("Коефіцієнт народжуваності",size=16, color="blue")
plt.show()
```

ВІЗУАЛІЗАЦІЯ

```python
df.query("Регіон == 'Миколаївська'")
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
date = df.loc[df['Регіон'] == 'Миколаївська', '1950':'2019'].columns.tolist()
k_per_year = df.loc[df['Регіон'] == 'Миколаївська', '1950':'2019'].values[0]
minimum_value = df.loc[df['Регіон'] == 'Миколаївська', '1950':'2019'].min().min()
plt.plot(date, k_per_year, label='Коефіцієнт народжуваності', linestyle="-", marker="o", color="#9966FF")
plt.xlabel('Дата', fontsize=15, color='blue')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='blue')
plt.title('Коефіцієнт народжуваності в Миколаївській області', fontsize=15)
plt.text(date[1], minimum_value, f'Мінімальний коефіцієнт = {minimum_value}', color="red", fontsize=12)
plt.xlim(0, 7)
plt.grid()
plt.show()
```

```python


labels = df['Регіон']
sizes = df.loc[:, '2019']

fig, ax = plt.subplots(figsize=(10, 10))
ax.pie(sizes, labels=labels, autopct='%.2f%%', startangle=90)
ax.axis('equal')
plt.title("Відсоток народжуванності 1960 року по областях", y=1.08)
plt.subplots_adjust(bottom=0.1)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

for index, row in df.iterrows():
    region = row['Регіон']
    k_per_year = row['1950':'2019'].values
    

    plt.plot(date, k_per_year, label=region, linestyle="-", marker="o")
    

plt.xlabel('Дата', fontsize=15, color='blue')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='blue')
plt.title('Коефіцієнт народжуваності в областях України', fontsize=15)
plt.xlim(0, 7)
plt.grid()
plt.legend(fontsize=6)
plt.show()
```
